```python
import numpy as np

x1 = np.random.normal(5, 7, size=1000)
x2 = np.random.normal(22.5, 2, size=1000)
y = x1 + x2 * 4 + 1
```

```python
df = pd.DataFrame(
    {"x1": x1,
     "x2": x2,
     "y": y})
```

```python
df.plot.scatter(x='x1', y='y')
```

```python
import seaborn as sns

sns.pairplot(df)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m = LinearRegression(fit_intercept=True)
m.fit(df[["x1", "x2"]], df["y"])
m.coef_, m.intercept_
```
