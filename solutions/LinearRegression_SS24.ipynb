### Try random slopes

```python
import random

best_mae = 99999999999999999999999999999
iter = 0

while best_mae > 3 and iter < 10_000:
    iter += 1
    # try out values between -50, +50
    a = random.random() * 1 - 0
    b = random.random() * 20 - 10

    ypred = a * df["x1"] + b  # Å· an estimate
    mae = sum(abs(ypred - df["y"])) / df.shape[0]

    if mae < best_mae:
        best_mae = mae
        print(f"a: {a:10.4f}   b: {b:10.4f}    MAE: {mae:10.4f}")
```

```python
# plot the result
from matplotlib import pyplot as plt

a = 0.4945  # 0.4995
b = 6.8865  # -2.9826
ypred = a * df["x1"] + b

plt.plot(df["x1"], df["y"], "b.")
plt.plot(df["x1"], ypred, "r-")
plt.grid()
```
