# Pretrained Neural Networks

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
```

```python
im = image.load_img(r'Desktop\tram.jpg', target_size=(224, 224))
im
```

```python
model = MobileNetV2(include_top=True)
```

```python
a = image.img_to_array(im)  # im is an image object -> numpy array
a.shape, a.min(), a.max()
```

```python
X = preprocess_input(a)
X = X.reshape(1, 224, 224, 3)
X.shape, X.min(), X.max()
```

```python
result = model.predict(X)
```

```python
result.shape, result.min(), result.max(), result.sum()
```

```python
decode_predictions(result)
```

```python
im = image.load_img(r'Desktop\fox.jpg', target_size=(224, 224))

a = image.img_to_array(im)
X = preprocess_input(a)
X = X.reshape(1, 224, 224, 3)

result = model.predict(X)
decode_predictions(result)
```
